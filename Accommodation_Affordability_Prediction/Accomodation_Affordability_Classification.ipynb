# PREPROCESSING

```python
#REQUIRED: Generate a list showing each feature and the the total number of null or missing values for that feature
df.isnull.sum()
```

```python
#REQUIRED: Fill the missing values for all numerical features with the mean. 
#NOTE: Marks will be allocated according to the efficiency of your code (as explained during your lectures)
num_cols=df.select_dtypes(include=['int64','float64']).columns
df[num_cols]=df[num_cols].fillna(df[num_cols].mean())
```

```python
#REQUIRED: Fill the missing values for all categorical features with the mode. 
#NOTE: Marks will be allocated according #to the efficiency of your code (as explained during your lectures)
cat_cols = df.select_dtypes(include=['object']).columns
for col in cat_cols:
    df[col].fillna(df[col].mode()[0], inplace=True)
```

```python
#REQUIRED: Generate a list showing each feature and the the total number of null or missing values for that feature
df.isnull().sum()
```

```python
#REQUIRED: What observations can you make about the list generated?

#Write your answer in this cell as comments below:

# 1. All missing values have been successfully handled since the
#    null count for every feature is now zero.

# 2. Numerical features were completed using the mean, which helps
#    preserve the overall distribution of the data.

# 3. Categorical features were completed using the mode, ensuring
#    that the most frequently occurring category was retained.

# 4. The dataset is now cleaner and more suitable for analysis
#    and machine learning tasks.

# 5. Handling missing values reduces the chances of errors during
#    model training and statistical computations.

# 6. Although missing values were filled, excessive imputation may
#    still introduce bias if many values were originally missing.
```

# Label Encoding

```python
#REQUIRED: Use LabelEncoder from sklearn to convert 'Day' and 'Type' categorical variables to numeric
#then show the dataframe.

from sklearn.preprocessing import LabelEncoder

# Create LabelEncoder object
le = LabelEncoder()

# Encode categorical columns
df['Day'] = le.fit_transform(df['Day'])
df['Type'] = le.fit_transform(df['Type'])

# Show dataframe
df
```

# One-Hot Encoding

```python
#REQUIRED: Use One-Hot Encoding on the 'City' feature. The final dataframe with the one-hot encoded 'City'
#feature must still be called df
df = pd.get_dummies(df, columns=['City'], dtype=int)
df
```

# MODEL TRAINING AND EVALUATION

```python
#REQUIRED: Split the dataframe into training features (X) and the target feature (y) 
X = df.drop('Affordability', axis=1)
y = df['Affordability']
```

```python
#REQUIRED: Use sklearn's train_test_split to split the dataset into x_train, x_test, y_train, y_test
#NOTE: use a test_size of 25% and random_state = 5
from sklearn.model_selection import train_test_split

# REQUIRED split
x_train, x_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, random_state=5
)
```

```python
#REQUIRED: Fit the LogisticRegression model to the data
from sklearn.linear_model import LogisticRegression

# Fit model
log_model = LogisticRegression(max_iter=1000)
log_model.fit(x_train, y_train)
```

```python
#REQUIRED: Display the confusion matrix for the LogisticRegression model
from sklearn.metrics import confusion_matrix

y_pred_log = log_model.predict(x_test)

cm_log = confusion_matrix(y_test, y_pred_log)
cm_log
```

```python
#REQUIRED: What observations can you make about the results of the confusion matrix? Explain the results.
#NOTE: Marks will be allocated for observations and explanations that go beyond just stating the obvious

#Write your answer in this cell as comments below:

# 1. True Positives and True Negatives indicate correct predictions,
#    showing how well the model distinguishes between classes.

# 2. False Positives indicate cases where the model incorrectly predicts
#    a positive class when it is actually negative.

# 3. False Negatives indicate missed positive cases, which may be critical
#    depending on the problem context.

# 4. A high number of diagonal values (TP + TN) indicates good model performance.

# 5. If off-diagonal values are high, the model is misclassifying data
#    and may require improvement or feature engineering.
```

```python
#REQUIRED: Display the classification report for the LogisticRegression model

from sklearn.metrics import classification_report

print(classification_report(y_test, y_pred_log))
```

```python
#REQUIRED: Display the accuracy score as a % for the LogisticRegression model in a user friendly manner
from sklearn.metrics import accuracy_score

acc_log = accuracy_score(y_test, y_pred_log)

print(f"Logistic Regression Accuracy: {acc_log * 100:.2f}%")
```

```python
#REQUIRED: Fit the RandomForestClassifier model to the data
from sklearn.ensemble import RandomForestClassifier

rf_model = RandomForestClassifier(random_state=5)
rf_model.fit(x_train, y_train)
```

```python
#REQUIRED: Display the confusion matrix for the RandomForestClassifier model
y_pred_rf = rf_model.predict(x_test)

cm_rf = confusion_matrix(y_test, y_pred_rf)
cm_rf
```

```python
#REQUIRED: Display the classification report for the RandomForestClassifier model
print(classification_report(y_test, y_pred_rf))
```

```python
#REQUIRED: Display the accuracy score as a % for the RandomForest model in a user friendly manner
acc_rf = accuracy_score(y_test, y_pred_rf)

print(f"Random Forest Accuracy: {acc_rf * 100:.2f}%")
```
