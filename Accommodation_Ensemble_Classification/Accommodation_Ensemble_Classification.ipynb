```python
n
import pandas as pd
import numpy as np
#import seaborn as sns
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score
from matplotlib import pyplot as plt
import matplotlib
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
```

```python
#REQUIRED: Keep the code below. You will just need to run this cell when needed

#Read in the dataset
df = pd.read_csv('Accommodation2.csv')

#Split the dataframe into training features (X) and the target feature (y) 
X = df.drop(columns=['Affordability'], axis=1) 
y = df['Affordability']   

#Use sklearn's train_test_split to split the dataset into x_train, x_test, y_train, y_test
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=5)
```

```python
#REQUIRED: Keep the code below. You will just need to run this cell when needed

#Fit the LogisticRegression model to the data
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(x_train, y_train)

#Display the accuracy of LogisticRegression model in a user friendly manner
print("LogisticRegression: ", model.score(x_test, y_test)*100, "%")
```

# Hard Voting

```python
#REQUIRED. Implement a VOTING CLASSIFIER using 'hard' voting
#Use the following information to complete this section:
#1. Use the folowing 3 estimators for the voting classifier :
#   GaussianNB(Parameters: none)
#   LogisticRegression (Parameters: solver="lbfgs", random_state=5)
#   DecisionTreeClassifier (Parameter: random_state=5)   
#2. Fit each of the classifiers and print the accuracy of all 4 models 
#   in a user friendly manner (Eg GaussianNB: 92.25 %)
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import VotingClassifier

gnb = GaussianNB()
lr = LogisticRegression(solver="lbfgs", random_state=5)
dt = DecisionTreeClassifier(random_state=5)

voting_clf = VotingClassifier(
    estimators=[
        ('gnb', gnb),
        ('lr', lr),
        ('dt', dt)
    ],
    voting='hard'
)

gnb.fit(x_train, y_train)
lr.fit(x_train, y_train)
dt.fit(x_train, y_train)
voting_clf.fit(x_train, y_train)

print("GaussianNB:", gnb.score(x_test, y_test)*100, "%")
print("LogisticRegression:", lr.score(x_test, y_test)*100, "%")
print("DecisionTreeClassifier:", dt.score(x_test, y_test)*100, "%")
print("VotingClassifier:", voting_clf.score(x_test, y_test)*100, "%")

```

# Stacking

```python
#REQUIRED. Implement a STACKING CLASSIFIER
#Use the following information to complete this section:
# 1. Base Estimators: DecisionTreeClassifier (Parameter: random_state=5) and SVC (Parameter: random_state=5)
#    Final Estimator: LogisticRegression(Parameters: None)
# 2. Fit the classifier and print the accuracy of the model
#    in a user friendly manner (Eg StackingClassifier 85.76 %)
#REQUIRED. Implement a STACKING CLASSIFIER
from sklearn.svm import SVC
from sklearn.ensemble import StackingClassifier

estimators = [
    ('dt', DecisionTreeClassifier(random_state=5)),
    ('svc', SVC(random_state=5))
]

stack_model = StackingClassifier(
    estimators=estimators,
    final_estimator=LogisticRegression()
)

stack_model.fit(x_train, y_train)

print("StackingClassifier:", stack_model.score(x_test, y_test)*100, "%")
```

# Bagging

```python
#REQUIRED. Implement a BAGGING CLASSIFIER
#Use the following information to complete this section:
# 1. Parameters: base_estimator is DecisionTreeClassifier (with class_weight = 'balanced')
#                max_samples=30500
# 2. Fit the classifier and print the accuracy of the model
#    in a user friendly manner (Eg BaggingClassifier 80.25 %)
#REQUIRED. Implement a BAGGING CLASSIFIER

from sklearn.ensemble import BaggingClassifier

base_model = DecisionTreeClassifier(class_weight='balanced')

bag_model = BaggingClassifier(
    estimator=base_model,
    max_samples=30500,
    random_state=5
)

bag_model.fit(x_train, y_train)

print("BaggingClassifier:", bag_model.score(x_test, y_test)*100, "%")
```

# Boosting

```python
#REQUIRED. Implement a BOOSTING CLASSIFIER using AdaBoostClassifier
#Use the following information to complete this section:
# 1. Parameters: base_estimator is DecisionTreeClassifier (with max_depth=8 and class_weight = 'balanced')
#    n_estimators=100, learning_rate = 0.5
# 2. Fit the classifier and print the accuracy of the model
#    in a user friendly manner (Eg AdaBoostClassifier 83.75%)

#REQUIRED. Implement a BOOSTING CLASSIFIER using AdaBoostClassifier

from sklearn.ensemble import AdaBoostClassifier

boost_base = DecisionTreeClassifier(
    max_depth=8,
    class_weight='balanced'
)

boost_model = AdaBoostClassifier(
    estimator=boost_base,
    n_estimators=100,
    learning_rate=0.5,
    random_state=5
)

boost_model.fit(x_train, y_train)

print("AdaBoostClassifier:", boost_model.score(x_test, y_test)*100, "%")
```
